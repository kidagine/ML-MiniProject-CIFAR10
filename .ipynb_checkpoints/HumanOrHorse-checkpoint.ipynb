## Horses Or Humans

# Setup

```python
# imports
import tensorflow as tf
from tensorflow import keras
import tensorflow_datasets as tfds
import sys
import sklearn
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

# The Horses Or Humans Dataset

This dataset contains 500 CGI images of horses and 500 CGI images of humans.

Horse (class 0)
Human (class 1)

We want to train a machine learning model to classify examples as either horse or human based on the images.
![title](horses_or_humans.png)

## Loading & Preparing

```python
(ds_train, ds_test) = tfds.load('horses_or_humans', split=['train', 'test'], shuffle_files=True, as_supervised=True)
```

## Preparing

```python
for images, labels in ds_test.take(2):  # only take first element of dataset
    plt.imshow(images)
    plt.axis('off')
    plt.show()
    print(labels)
```

## Training

```python
model = keras.models.Sequential()
model.add(keras.layers.Flatten(input_shape=[28, 28]))

model.add(keras.layers.Dense(300, activation = "relu"))
model.add(keras.layers.Dense(100, activation = "relu"))

model.add(keras.layers.Dense(10, activation = "softmax"))
```

## Evaluation

```python
model.summary()
```

```python
model.compile(loss="sparse_categorical_crossentropy", optimizer="sgd", metrics=["accuracy"])
```

```python
history = model.fit(ds_train, epochs=10, batch_size=32, validation_data = ds_train)
```
